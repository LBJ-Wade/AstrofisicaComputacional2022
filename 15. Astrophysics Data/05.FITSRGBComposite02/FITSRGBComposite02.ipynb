### Resumen

En este cuaderno utilizaremos `astropy` para leer un conjunto de archivos .fits y visualizar una versión compuesta, a partir de información de diferentes filtros, de una imágen astronómica.

---

## 1. Los Archivos  .FITS 

En este segundo ejemplo de composición de imagenes, utilizaremos los archivos .fits de la galaxia NGC123 (M81) obtenidos de la base de datos SDSS,

https://dr12.sdss.org/fields/name?name=M81

Se descargaran las imagenes en las bandas *i*, *r* y *g*. Para la información de las bandas en el SDSS vea [aquí](https://www.sdss.org/instruments/camera/)



```python
import numpy as np
from matplotlib import pyplot as plt

from astropy.io import fits
from astropy.visualization import make_lupton_rgb

hduli = fits.open('data/NGC123i.fits')
hdulr = fits.open('data/NGC123r.fits')
hdulg = fits.open('data/NGC123g.fits')
hdulu = fits.open('data/NGC123u.fits')

i = hduli[0].data
r = hdulr[0].data
g = hdulg[0].data
u = hdulu[0].data
```

## 2. Visualización de la Imagen


```python
rgb_default = make_lupton_rgb(i, r, g)

plt.figure(figsize=(10,10))
plt.imshow(rgb_default, origin='lower')
plt.show()
```

```python
rgb_default = make_lupton_rgb(i, r, g, stretch=0.8, Q=18)

plt.figure(figsize=(10,10))
plt.imshow(rgb_default, origin='lower')
plt.show()
```
